```python
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup
headers = {
    "User-Agent": "Mozilla/5.0 (SeminarScraper/1.0; +https://example.org/)",
    "Accept": "text/html,application/xhtml+xml",
    "Accept-Language": "ru,en;q=0.8",
}

session = requests.Session()
session.headers.update(headers)
BASE = "https://ru.wikipedia.org"

r1 = session.get("https://ru.wikipedia.org/wiki/Служебная:Случайная_страница", timeout=20)
soup1 = BeautifulSoup(r1.text, "lxml")
links1 = soup1.find_all("a", href=True)
link1 = links1[5]['href']
article_url1 = urljoin(BASE, link1)
print("Ссылка на первую статью:", r1.url)
title1 = soup1.find("h1").get_text(strip=True)
print("Название первой статьи:", title1)

r2 = requests.get(article_url1, headers=headers, timeout=20)
soup2 = BeautifulSoup(r2.text, "lxml")
links2 = soup2.find_all("a", href=True)
link2 = links2[5]['href']
article_url2 = urljoin(BASE, link2)
print("Ссылка на вторую статью:", article_url1)
title2 = soup2.find("h1").get_text(strip=True)
print("Название второй статьи:", title2)

r3 = requests.get(article_url2, headers=headers, timeout=20)
soup3 = BeautifulSoup(r3.text, "lxml")
print("Ссылка на третью статью:", article_url2)
title3 = soup3.find("h1").get_text(strip=True)
print("Название третьей статьи:", title3)
```
